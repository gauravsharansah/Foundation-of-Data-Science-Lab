```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

n = 1000

# ID
ID = np.arange(1, n+1)

# Age with missing values
Age = np.random.randint(18, 70, size=n).astype(float)
Age[np.random.choice(n, size=int(0.05*n), replace=False)] = np.nan

# Gender with missing values
Gender = np.random.choice(['Male', 'Female', 'Other'], size=n)
Gender[np.random.choice(n, size=int(0.03*n), replace=False)] = None

# Education with missing values
Education = np.random.choice(['High School', 'Bachelors', 'Masters', 'PhD'], size=n)
Education[np.random.choice(n, size=int(0.04*n), replace=False)] = None

# Salary with outliers
Salary = np.random.normal(70000, 20000, size=n).astype(float)
Salary[np.random.choice(n, size=int(0.01*n), replace=False)] *= 3  # Outliers
Salary = np.round(Salary, 2)

# Experience with missing values
Experience = np.random.randint(0, 50, size=n).astype(float)
Experience[np.random.choice(n, size=int(0.05*n), replace=False)] = np.nan

# Department
Department = np.random.choice(['IT', 'HR', 'Finance', 'Marketing', 'Operations'], size=n)

# Remote_Work with missing
Remote_Work = np.random.choice(['Yes', 'No'], size=n)
Remote_Work[np.random.choice(n, size=int(0.02*n), replace=False)] = None

# Joining Year with missing
Joining_Year = np.random.randint(1990, 2026, size=n).astype(float)
Joining_Year[np.random.choice(n, size=int(0.03*n), replace=False)] = np.nan

# Performance Score with outliers
Performance_Score = np.random.randint(1, 11, size=n).astype(float)
Performance_Score[np.random.choice(n, size=int(0.01*n), replace=False)] *= 2  # Outliers
Performance_Score = np.clip(Performance_Score, 1, 10)

# Certifications with missing
Certifications = np.random.randint(0, 11, size=n).astype(float)
Certifications[np.random.choice(n, size=int(0.1*n), replace=False)] = np.nan

# Attrition
Attrition = np.random.choice(['Yes', 'No'], size=n)

# Create DataFrame
df = pd.DataFrame({
    'ID': ID,
    'Age': Age,
    'Gender': Gender,
    'Education': Education,
    'Salary': Salary,
    'Experience': Experience,
    'Department': Department,
    'Remote_Work': Remote_Work,
    'Joining_Year': Joining_Year,
    'Performance_Score': Performance_Score,
    'Certifications': Certifications,
    'Attrition': Attrition
})

# Save to CSV
df.to_csv('employee_data.csv', index=False)
print("Dataset created successfully with 1000 rows and 12 features.")
```
