```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Line Chart

```python
df = pd.read_csv("Health_heart_experimental.csv")
# create a line chart, age on x-axis, blood-pressure on y-axis

# Group by age and indication, then calculate mean BP
avg_bp = df.groupby(['age', 'indication'])['SysBP'].mean().reset_index()
bp_0 = avg_bp[avg_bp['indication'] == 0]
bp_1 = avg_bp[avg_bp['indication'] == 1]
plt.figure(figsize=(10,6))
plt.plot(bp_0['age'], bp_0['SysBP'], marker='o', label='Healthy')
plt.plot(bp_1['age'], bp_1['SysBP'], marker='s', label='Clinical Risk')

plt.xlabel("Age (years)")
plt.ylabel("Average Systolic BP (mmHg)")
plt.title("Average Blood Pressure vs Age by Indication")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
# create a line chart, age on x-axis, blood-pressure on y-axis

# Group by age and indication, then calculate mean BP
avg_bp = df.groupby(['age', 'indication'])['DiaBP'].mean().reset_index()
bp_0 = avg_bp[avg_bp['indication'] == 0]
bp_1 = avg_bp[avg_bp['indication'] == 1]
plt.figure(figsize=(10,6))
plt.plot(bp_0['age'], bp_0['DiaBP'], marker='o', label='Healthy')
plt.plot(bp_1['age'], bp_1['DiaBP'], marker='s', label='Clinical Risk')

plt.xlabel("Age (years)")
plt.ylabel("Average Diastolic BP (mmHg)")
plt.title("Average Blood Pressure vs Age by Indication")
plt.legend()
plt.grid(True)
plt.show()
```

This line chart shows that the average blood pressure of every age group for healthy and clinical risk person are a kind of similar with most of peaks occupied by clinical risk person. This could be because clinical risk person have high BP because of hypersion stage1/2 and healthy people have high BP because of social life, anger,etc. And from this line chart it can be insighted that Blood Pressure does not have vital impact in deciding the heart disorder of a person.

### Pie-Chart

```python

df["indication_label"] = df["indication"].map({0: "Healthy", 1: "Clinical_Risk"})
df['indication_label'].value_counts().plot(kind='pie',autopct='%0.2f')
```

This shows that similar no of participants were healthy and at risk stage due to one or more causes among blood pressure, heart rate, weight abnormality or BMI. This shows clearly that anyway the no of healthy people are more than risky people which shows that many of them are health conscious too.

### Bar Graph

```python
# BMI classification
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Healthy"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI_Category"] = df["BMI"].apply(bmi_category)
bmi_counts = df["BMI_Category"].value_counts()


plt.bar(bmi_counts.index, bmi_counts.values)
plt.xlabel("BMI Category")
plt.ylabel("Number of Persons")
plt.title("Distribution of BMI Categories")

plt.ylim(bottom=0)
plt.xticks(rotation=45)

plt.show()
```

This bar graph supports the insights from pie chart as its clear that more that 40% of participants are obese which could be the most primariry reason for heart related problems.And, most probably above 90% of obese participants and few overweight participants make up the clinical risk category of people in this experiment.Hence, weight is an important factor for a healthy people.It provides the insight that one must have a proper care about diet,weight which directly influences,BMI i.e. Body Mass Index.

### Box Plot

```python

df["indication_label"] = df["indication"].map({0: "Healthy", 1: "Clinical_Risk"})
df["sex_label"] = df["sex"].map({0: "Male", 1: "Female"})
sns.boxplot(x='sex_label',y='heightCm',hue='indication_label',data=df)
```

This box plot shows that the healthy participants has a height of more than 125cm and also with the fact that less people are scattered in height of 125 cm to 150cm. This shows that height plays important role in healthiness because the normal height of healthy people always ranges between 135 cm to 175cm except abnormality cases.This study insights that the people with lower height are more prone to heart disorders for such people the cause could be depression and social impact, and for those people which are in good height range have heart disorder because of lifestyle and diet.

### Scatter Plot

```python
plt.scatter(df['heightCm'], df['weightKg'], color='green', alpha=0.1, s=2)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight Scatter Plot")
plt.grid(True)
plt.show()
```

Since , the participants with good height also have low weights such as 50kg, 60 kg, they are underweight category with less bmi, and also participants with lower height has more weight which shows that some of obese participants are of smaller height.
